```python
import math

def mexpx(ex, i):
    result = 0
    for j in range(i):
        result += ex**j / math.factorial(j)
    return result
```

```python
import pandas as pd

df = pd.DataFrame()
xlist = []
ylist = []
yrlist = []

for n in range(30):
    x = ((n-10)/10)
    xlist.append(x)
    ylist.append(mexpx(x, 5))
    yrlist.append(math.exp(x))

df = pd.DataFrame()
df['x'] = xlist
df['y'] = ylist
df['yr'] = yrlist

df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6), dpi=200)
sns.lineplot(data=df, x='x', y='y')
sns.lineplot(data=df, x='x', y='yr')
```
