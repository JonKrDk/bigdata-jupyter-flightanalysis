```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

```python
with open('flights.json','r') as f:
    data = json.loads(f.read())
```

```python
df = pd.json_normalize(data, record_path =['response'])
```

```python
df
```

```python
plt.figure(figsize=(12,6), dpi=200)
sns.scatterplot(data=df, x='alt', y='speed', hue='v_speed', palette='plasma')
```

```python
plt.figure(figsize=(12,6), dpi=200)
sns.scatterplot(data=df, x='v_speed', y='speed', hue='alt', palette='hsv')
```

```python
plt.figure(figsize=(12,6), dpi=200)
sns.scatterplot(data=df, x='lng', y='lat', hue='speed')
```

```python
plt.figure(figsize=(12,6), dpi=200)
sns.scatterplot(data=df, x='lng', y='lat', hue='alt', style='status')
```

```python
sns.displot(data=df, x='speed', bins=50, height=8, aspect=3, kde=True)
```
