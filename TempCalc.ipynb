```python
import pandas as pd
```

```python
df = pd.read_csv(filepath_or_buffer="c:\\Users\\jon\\desktop\\data1.csv", sep=';', decimal=',')
```

```python
df
```

```python
dfnt = df.drop("Timestamp", axis=1)
```

```python
dfnt.set_index(" Year")
```

```python
dfgr = dfnt.groupby(" Year").mean()
```

```python
dfgr
```

```python
dfgr["Average"] = dfgr.mean(axis=1)
```

```python
dfgr
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12,6), dpi=200)
#sns.scatterplot(data=dfgr, x=dfgr.index, y='Average')
sns.lmplot(data=dfgr, x=dfgr.index.astype(float), y='Average')
```
